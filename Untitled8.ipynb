```python
import requests
import pandas as pd
import time

def extraer_historial_mundiales(años):
    todos_los_partidos = []

    for año in años:
        print(f"⏳ Extrayendo datos del año {año}...")
        # URL de ESPN para resultados históricos
        url = f"https://site.api.espn.com/apis/site/v2/sports/soccer/fifa.world/scoreboard?dates={año}"

        try:
            res = requests.get(url)
            data = res.json()

            for evento in data.get('events', []):
                comp = evento['competitions'][0]
                fecha = evento['date']
                equipo1 = comp['competitors'][0]['team']['displayName']
                score1 = comp['competitors'][0]['score']
                equipo2 = comp['competitors'][1]['team']['displayName']
                score2 = comp['competitors'][1]['score']

                todos_los_partidos.append({
                    'fecha': fecha,
                    'local': equipo1,
                    'goles_local': score1,
                    'visitante': equipo2,
                    'goles_visitante': score2,
                    'mundial': año
                })
        except Exception as e:
            print(f"Error en año {año}: {e}")

        time.sleep(1) # Pausa para no saturar la API

    df = pd.DataFrame(todos_los_partidos)

    # Guardar en Drive
    csv_path = '/content/drive/MyDrive/Mundial2026/data/historial_partidos.csv'
    df.to_csv(csv_path, index=False)
    print(f"💾 ¡Éxito! Guardados {len(df)} partidos en {csv_path}")
    return df

# Ejecutamos para los últimos 3 mundiales
df_historia = extraer_historial_mundiales([2014, 2018, 2022])
df_historia.head()
```
